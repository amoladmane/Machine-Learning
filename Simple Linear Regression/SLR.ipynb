# Read dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("50_Startups.csv")
df.head()
```

```python
df.describe().T
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(data=df, x='R&D Spend', y='Profit')
plt.show()
```

# Seprate Profit and RND 

```python
df.columns
```

```python
X = df[['R&D Spend']]
Y = df[['Profit']]
```

# Creating Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
sns.regplot(data=df, x='R&D Spend' , y='Profit' , line_kws={'color':'blue'})
plt.show()
```

## Estimating out of Sample Data

```python
rnd =[[80000]]
pred_profit = model.predict(rnd)
```

```python
pred_profit
```

# R2 Score

```python
model.score(X,Y)
```

## Predict result on training data

```python
pred = model.predict(X)
pred[0:5]
```

```python
Y.head()
```

## Evaluating model 
1. MSE - measn sqaured error
2.RMSE - root mean sqaured error
3. MAE - mean absolute erro
4. MAPE - mean absolute percentage error

```python
# R2 score
model.score(X,Y)
```

```python
from sklearn.metrics import mean_squared_error
ms = mean_squared_error(Y, ypred)
ms
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(Y,ypred)
```

```python
mape
```
